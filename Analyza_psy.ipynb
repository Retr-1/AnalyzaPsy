```python
print('mnozstvo plemien:', len(dogs["breed_group"].unique()))
a = dogs.groupby("breed_group")["Aggressive_people"].mean(numeric_only=True)
print("jednotlive plemena:")
display(a.sort_values())

# dogs_na_tuto_tabulku = dogs.loc[:, ["Aggressive_people", 'Aggressive_dogs_samegender', 'Aggressive_dogs_oppositegender']]
# dogs_na_tuto_tabulku["Aggressive_dogs_samegender"] = dogs_na_tuto_tabulku["Aggressive_dogs_samegender"]/len(dogs_na_tuto_tabulku["Aggressive_dogs_samegender"])
# dogs_na_tuto_tabulku["Aggressive_people"] = dogs_na_tuto_tabulku["Aggressive_people"]/len(dogs_na_tuto_tabulku["Aggressive_people"])
# dogs_na_tuto_tabulku["Aggressive_people"] = dogs_na_tuto_tabulku["Aggressive_dogs_oppositegender"]/len(dogs_na_tuto_tabulku["Aggressive_dogs_oppositegender"])

print('mnozstvo samcov:', len(dogs.query("sex=='male'")))
print('mnozstvo samic:', len(dogs.query("sex=='female'")))
# plot, axes = plt.subplots(2, 3, figsize=(15, 8), sharey=True)

male_count = len(dogs.query("sex=='male'"))
female_count = len(dogs.query("sex=='female'"))
plot, axes = plt.subplots(3, 2, figsize=(8, 10), sharey=True)


# Male dogs
sns.countplot(data=dogs.query("sex=='male'"), x='Aggressive_people', ax=axes[0][0])
axes[0][0].set(title='Male: Aggression Toward People', xlabel='Aggression Score', ylabel='Count')

sns.countplot(data=dogs.query("sex=='male'"), x='Aggressive_dogs_samegender', ax=axes[1][0])
axes[1][0].set(title='Male: Same-Gender Aggression', xlabel='Aggression Score', ylabel='Count')

sns.countplot(data=dogs.query("sex=='male'"), x='Aggressive_dogs_oppositegender', ax=axes[2][0])
axes[2][0].set(title='Male: Opposite-Gender Aggression', xlabel='Aggression Score', ylabel='Count')

# Female dogs
sns.countplot(data=dogs.query("sex=='female'"), x='Aggressive_people', ax=axes[0][1])
axes[0][1].set(title='Female: Aggression Toward People', xlabel='Aggression Score', ylabel='Count')

sns.countplot(data=dogs.query("sex=='female'"), x='Aggressive_dogs_samegender', ax=axes[1][1])
axes[1][1].set(title='Female: Same-Gender Aggression', xlabel='Aggression Score', ylabel='Count')

sns.countplot(data=dogs.query("sex=='female'"), x='Aggressive_dogs_oppositegender', ax=axes[2][1])
axes[2][1].set(title='Female: Opposite-Gender Aggression', xlabel='Aggression Score', ylabel='Count')

plot.tight_layout()
plt.show()
```
